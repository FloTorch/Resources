# üè• Building an Intelligent Medical Assistant with Flotorch Autogen

This notebook demonstrates how to build a smart, memory-enabled medical assistant using the `FlotorchAutogenAgent`. Our goal is to create a **Medical Consultation Agent** that intelligently gathers patient information and provides healthcare guidance while remembering patient history across sessions.

### Objective

The agent leverages the `FlotorchAutogenAgent` client to connect to an agent pre-configured on the Flotorch platform. This remote agent is equipped with **Memory capabilities** to perform several key functions:

* **Collect Patient Information**: Gather essential details (Name, Age, Gender)
* **Understand Medical Needs**: Listen to symptoms and health concerns
* **Provide Recommendations**: Offer dietary advice, lifestyle suggestions, and treatment guidance
* **Remember Patient History**: Store and retrieve patient data across sessions using Mem0

### Key Advantage: Agent + Memory Integration

This notebook's primary benefit is the **seamless integration of agent intelligence with persistent memory**. All patient interactions are stored automatically, enabling the agent to:
- Recognize returning patients instantly
- Recall previous consultations and recommendations
- Provide personalized care based on medical history
- Build long-term patient relationships

The agent configuration and memory provider are managed entirely on the Flotorch platform UI, allowing you to update behavior and storage settings without ever changing this notebook's code.

## Prerequisites

Before running this notebook, you must configure two components in the Flotorch console (https://console.flotorch.cloud/):

1. **Agent Configuration**: Create an agent named `medi-assist` (or update the `AGENT_NAME` variable below)

## 3. Import Required Libraries

### Purpose
Import all required components for building the medical assistant with Autogen framework.

### Key Components
- **FlotorchAutogenAgent**: Connects to remotely configured agent on Flotorch platform
- **FlotorchAutogenSession**: Manages conversation state and session persistence
- **FlotorchAutogenMemory**: Handles persistent patient data storage across sessions
- **RoundRobinGroupChat**: Orchestrates multi-agent conversations in round-robin fashion
- **MemoryContent, MemoryMimeType**: Utilities for formatting and handling memory content
- **Console**: Provides interactive UI for displaying agent conversations


````python
# Import necessary libraries
from flotorch.autogen.agent import FlotorchAutogenAgent
from flotorch.autogen.sessions import FlotorchAutogenSession
from flotorch.autogen.memory import FlotorchAutogenMemory
from autogen_agentchat.teams import RoundRobinGroupChat
from autogen_core.memory import MemoryContent, MemoryMimeType
from autogen_agentchat.ui import Console

print("Imported necessary libraries successfully")
````


## 4. Initialize Memory Service

### Purpose
Set up the infrastructure for:
1. Set up the `Mem0 memory provider` to enable persistent patient data storage across sessions.

### Memory Service
The memory service enables the agent to:
- Remember patient details across sessions
- Retrieve previous diagnoses and treatments
- Recognize returning patients automatically

````python
session_memory = FlotorchAutogenSession(
    api_key=FLOTORCH_API_KEY,
    base_url=FLOTORCH_BASE_URL
)


external_memory = FlotorchAutogenMemory(
    api_key=FLOTORCH_API_KEY,
    base_url=FLOTORCH_BASE_URL,
    name = MEMORY_PROVIDER,
    user_id = USER_ID,
    app_id  = APP_ID
)

print("Initialized external_memory")
````

## 5. Initialize Session Service

### Purpose
Set up the infrastructure for:
1. `Session Management` : Tracks current conversation context

### Session Service
The session service manages:
- Current conversation flow
- Message history for context
- Turn-by-turn interaction state

````python
session_memory = FlotorchAutogenSession(
    api_key=FLOTORCH_API_KEY,
    base_url=FLOTORCH_BASE_URL
)

print("Initialized session_memory")
````

## 6. Initialize Medical Agent

Connect to the remotely configured **medical assistant agent** with memory enabled.

### Purpose
Connect to the remotely configured medical assistant agent.

#### Key Features
- Collects patient information systematically (Name ‚Üí Age ‚Üí Gender ‚Üí Symptoms)
- Provides empathetic healthcare guidance
- Offers dietary and lifestyle recommendations
- Stores all consultations in structured format
- Recognizes and personalizes care for returning patients

````python
flotorch_client = FlotorchAutogenAgent(
    agent_name =AGENT_NAME,
    api_key=FLOTORCH_API_KEY,
    base_url=FLOTORCH_BASE_URL,
    memory = [external_memory],
    model_context = session_memory,

)

agent = flotorch_client.get_agent()

print("Fetched agent is initialized")
````

## 7. Create Agent Team with RoundRobinGroupChat

### Purpose
Assemble the agent into a team structure for orchestrated conversation management.

### What is RoundRobinGroupChat?
A team orchestration pattern that manages multi-agent conversations in a round-robin fashion. Even with a single agent, it provides structured conversation flow control.

### Configuration Details
- **participants**: List of agents participating in the conversation (our medical assistant)
- **max_turns**: Maximum number of conversation turns per interaction (set to 1 for direct responses)

### Benefits
- **Structured Flow**: Organizes agent responses in a predictable pattern
- **Scalability**: Easy to add more agents if needed (e.g., specialist consultants)
- **Control**: Limits conversation length with `max_turns` parameter
- **Integration**: Seamlessly works with memory and session management

````python
team = RoundRobinGroupChat(
        participants=[agent],
        max_turns=1
    )
print(f"Agent and Task created successfully: {agent.name}")
````

## 8. Interactive Chat Interface

### Purpose
Start an interactive conversation with the medical assistant.

### Features
- **Continuous Conversation**: Chat until you type "exit"
- **Automatic Memory**: Patient data saved automatically
- **Session Persistence**: Conversation context maintained throughout

### Usage Instructions
1. Run the cell below
2. Enter your messages when prompted
3. Type `exit` to end the conversation

### Example Interaction Flow
```
You: Hello
Dr. Hello! How can I assist you with your health today? Before we get started, may I have your name, age, and gender?

You: My name is John, I'm 35, male
Dr. AI: Thank you, John. What brings you here today?

You: I have a headache
Dr. AI: I understand. Can you describe the headache? When did it start?...
```

### Behind the Scenes
- Agent collects: Name, Age, Gender
- Then gathers: Symptoms, Medical History, Medications, Allergies
- Stores everything in structured memory format
- Next visit: Automatically recognizes "John" and retrieves history

````python
while True:
    user_input = input("user: ")

    if user_input.lower().strip() == "exit":
        break
    
    # Use this for normal agent.
    stream = team.run_stream(task = user_input)
    
    # Use this if you configured output schema to the agent.
    stream = agent.run_stream(task = user_input)
    result = await Console(stream)
    #  --- Add user input to memory ---
    await external_memory.add(MemoryContent(content=user_input, mime_type=MemoryMimeType.TEXT))

    # --- Add assistant response(s) to memory ---
    if hasattr(result, "messages"):
        for msg in result.messages[-1]:
            # Only save text messages (type == "TextMessage")
            if getattr(msg, "type", None) == "TextMessage":
                
                await external_memory.add(MemoryContent(content=getattr(msg, "content", ""), mime_type=MemoryMimeType.TEXT))
````
